ANALYZE EXISTENCE OF GLOBAL WARMING

```python
global_temp = pd.read_csv(r"C:\Users\Derek Egenti\Geospatial\Global Warming\GlobalTemperatures.csv")

global_temp.head(5)
```

We want to extract the year from the date column

```python
global_temp['dt'][0].split('-')[0]
```

We create a function to split the year from the date and return only the year.

```python
def fetch_year(date):
    return date.split('-')[0]
```

Next, we create a new column to pass the result of running the fuction through every row.

```python
global_temp['years'] = global_temp['dt'].apply(fetch_year)
```

```python
global_temp.head(4)
```

Now, we want to fin dthe average tempearture and average temperature uncertainty for each year. We will use the groupby and aggregator functions.

```python
data = global_temp.groupby('years').agg({'LandAverageTemperature': 'mean', 'LandAverageTemperatureUncertainty': 'mean'}).reset_index()
```

```python
data.head(5)
```

If you want to define 2 new columns, Uncertainty_top(LandAverageTemperature + LandAverageTemperatureUncertainty) and 
uncertainty_bottom(LandAverageTemperature - LandAverageTemperatureUncertainty)

```python
data['Uncertainty_Top'] = data['LandAverageTemperature'] + data['LandAverageTemperatureUncertainty']
data['Uncertainty_Bottom'] = data['LandAverageTemperature'] - data['LandAverageTemperatureUncertainty']
```

```python
data.head(4)
```

Now, we plot.

```python
fig = px.line(data, x = 'years', y = ['LandAverageTemperature', 'Uncertainty_Top', 'Uncertainty_Bottom'], title= 'Average Land temperature in the World')

fig.show()
```

ANALYSE AND VISUALIZE AVERAGE TEMPERATURE IN EACH SEASON

```python
global_temp.head()
```

To analyse seasons, were going to have to create a season column. We can do this by geeting month out of the 'dt' column.


First, we convert the dt column to a datetime column.

```python
global_temp['dt'] = pd.to_datetime(global_temp['dt'])
```

```python
global_temp['month'] = global_temp['dt'].dt.month
```

```python
global_temp.head()
```

```python
global_temp['month'].unique()
```

Now, we create a function to seperate the months into seasons

```python
def get_seasons(month):
    
    if month >=3 and month <=5:
        return 'spring'
    elif month >=6 and month <=8:
        return 'summer'
    elif month >=9 and month <=11:
        return 'autumn'
    else:
        return 'winter'
```

```python
global_temp['season'] = global_temp['month'].apply(get_seasons)

global_temp.head(3)
```

```python
years = global_temp['years'].unique()
```

Now, we write a code to partition the years and find the mean of the temperatures 

```python
spring_temps = []
summer_temps = []
autumn_temps = []
winter_temps = []
```

```python
for year in years:
    current_df = global_temp[global_temp['years'] == year] ##This is a filter used to chesck if the value stored in the 'year' variable is present in the column 'years'. 
    spring_temps.append(current_df[current_df['season']=='spring']['LandAverageTemperature'].mean())
    summer_temps.append(current_df[current_df['season']=='summer']['LandAverageTemperature'].mean())
    autumn_temps.append(current_df[current_df['season']=='autumn']['LandAverageTemperature'].mean())
    winter_temps.append(current_df[current_df['season']=='winter']['LandAverageTemperature'].mean())
```

create a dataframe with all the years and all the average temperatures

```python
season = pd.DataFrame()
```

```python
season['year'] = years
season['spring_temps'] = spring_temps
season['summer_temps'] = summer_temps
season['autumn_temps'] = autumn_temps
season['winter_temps'] = winter_temps
```

```python
season.head()
```

```python
season.columns
```

```python
fig_1 = px.line(season, x = 'year', y = ['spring_temps', 'summer_temps', 'autumn_temps', 'winter_temps'], title= 'Average Land temperature of Seasons Per Year')

fig_1.show()
```

```python
cities.head(4)
```

```python
India = cities[cities['Country'] == 'India']

# This is called Boolean Indexing. The filter here is cities['Country'] == 'India' and it is passed to the dataframe cities[].
```

```python
India.head(4)
```

```python
India.shape
```

We define the major indian cities by population, so we extract the cities with the highest population density.

```python
cities = ['New Delhi', 'Bangalore', 'Hyderabad', 'Pune', 'Madras', 'Varanasi', 'Gurgaon']
```

```python
India['City'].isin(cities) # This is the filter, so we pass it into the dataframe India[]

cities = India[India['City'].isin(cities)]

cities.head(4)
```

We want to remove the N and the E from the latitude and longitude values, respectively

```python
cities['Latitude'] = cities['Latitude'].str.strip('N')
cities['Longitude'] = cities['Longitude'].str.strip('E')
```

```python
cities.head(4)
```

We want to fetch the month column from the data next. First thing is to convert the dt column to datetype.

```python
cities['dt'] = pd.to_datetime(cities['dt'])
```

```python
cities['Month'] = cities['dt'].dt.month
```

```python
cities.head(4)
```

Next, we want to get a dataframe that is grouped by the basis of month and city, and has an average temperature column.

```python
cities_temp = cities.groupby(['Month', 'City'])['AverageTemperature'].mean().to_frame().reset_index()

cities_temp.head()
```

Then we do a simple column renaming

```python
cities_temp.columns = ['month', 'City', 'Mean_temp']

cities_temp.head()
```

We need the latitudes and longitudes, so we merge the cities dataframe with the cities_temp dataframe

```python
df = cities_temp.merge(cities, on = 'City')
df.head()
```

As we can see, we need to drop the duplicates.

```python
data = df.drop_duplicates(subset = ['month', 'City'])

data.head()
```

```python
data.columns
```

```python
data2 = data[['month', 'City', 'Mean_temp', 'Country', 'Latitude', 'Longitude']]

data2.head()
```

Now, we want to plot a heatmap that has 3 main columns, so the process of plotting it will be different from the other ones.

```python
import plotly.graph_objs as go

data = [go.Heatmap(x=data2['month'],
           y=data2['City'],
           z=data2['Mean_temp'])]
```

```python
layout = go.Layout(title = 'Avg Temp of Major Indian cities by Month')
```

```python
fig = go.Figure(data=data, layout  = layout)
fig.show()
```
