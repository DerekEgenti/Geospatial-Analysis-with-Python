```python
from geopy.geocoders import Nominatim


geolocator = Nominatim(user_agent='app', timeout=None)
```

```python
rest_loc = pd.DataFrame()
```

```python
rest_loc['Name'] = df['location'].unique()

rest_loc
```

```python
lat = []
lon = []

for name in rest_loc['Name']:
    location = geolocator.geocode(name)
    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
        
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
print(lat)
```

```python
print(lon)
```

```python
rest_loc['Lat'] = lat
rest_loc['Lon'] = lon

rest_loc
```

## Dealing with missing values

We want to find out which restaurants have the highest ratings, so we are going to work with the rate column.
It will probably have some missing values.

```python
df['rate'].isnull().sum()
```

```python
df['rate'].isnull().sum()/len(df)*100 #percentage of dat missing
```

We are going to drop the missing values

```python
df.dropna(subset = ['rate'], inplace = True)
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].unique() ## To get all the unique entries in the mentioned column.
```

```python
df.replace('NEW', '0', inplace = True)
df.replace('-', '0', inplace = True)
```

```python
df['rate'].unique()
```

```python
df['rating'] = df['rate'].str.replace('/5', '')
```

```python
df['rating']
```

```python
df['rating'].unique()
```

The entries are in string format. We want to convert then to float.

```python
df['rating'] = df['rating'].astype(float)
df['rating']
```

```python
df['rating'].dtype
```

```python
df['rating'].unique()
```

```python
df.columns
```

```python
grp_df = df.groupby(['location'], as_index = False).agg({'rating': 'mean', 'name': 'size'})

grp_df.columns = ['Name', 'avg_rating', 'count']

grp_df
```

As we can see from above, some locations have little number of restaurants, so we filter them out.
We use a threshold of 400 restaurants minimum. Meaning only locations with more than 400 restaurants will be kept.

```python
grp_df['count']>400
```

```python
temp_df = grp_df[grp_df['count']>400]
temp_df
```

Now, we carry the lat and lon data from the rest_loc dataframe. We can do this by merging the temp_df and rest_loc using the 'Name' column as the common column.

```python
Ratings_locations = temp_df.merge(rest_loc, on='Name')
Ratings_locations
```

Now, we construct a heatmap, using the average ratings a definer.

```python
rate_basemap = Generate_basemap()
```

```python
HeatMap(Ratings_locations[['Lat', 'Lon', 'avg_rating']]).add_to(rate_basemap)
```
